## Calculating some batting averages

```python
df_batters['Avg_Hits_PerGame'] = df_batters['Hits'] / df_batters['G_x']
df_batters['Avg_HR_PerGame'] = df_batters['HR'] / df_batters['G_x']
df_batters['Avg_Strikeouts_PerGame'] = df_batters['Strikeouts'] / df_batters['G_x']
df_batters['Avg_Hit_by_Pitch_PerGame'] = df_batters['Hit_by_Pitch'] / df_batters['G_x']
df_batters['Avg_Runs_PerGame'] = df_batters['Runs'] / df_batters['G_x']
df_batters['Avg_RBI_PerGame'] = df_batters['RBI'] / df_batters['G_x']
df_batters['Avg_Grounded_into_DP_PerGame'] = df_batters['GIDP'] / df_batters['G_x']
df_batters['Avg_Doubles_PerGame'] = df_batters['Doubles'] / df_batters['G_x']
df_batters['Avg_Triples_PerGame'] = df_batters['Triples'] / df_batters['G_x']
#df_batters.drop(columns=['player_name'], inplace=True)
df_batters
```

```python
df_batters['Singles'] = df_batters['Hits'] - (df_batters['Doubles'] + df_batters['Triples'] + df_batters['HR'])
df_batters['Avg_Singles_PerGame'] = df_batters['Singles'] / df_batters['G_x']

# batting average
df_batters['BA'] = df_batters['Hits'] / df_batters['At_Bats']
# Slug %
df_batters['SLG'] = (df_batters['Singles'] + 2*df_batters['Doubles'] + 3*df_batters['Triples'] + 4*df_batters['HR']) / df_batters['At_Bats']
# ISO is similar to slugging score: 'SLG', but it represents a batters raw power by only calculating big hits
df_batters['ISO'] = (df_batters['Doubles'] + 2*df_batters['Triples'] + 3*df_batters['HR']) / df_batters['At_Bats']

# a lot of 'bb_type' values are null, because no hit was recorded on the pitch
df_batters['bb_type'].fillna('no_hit', inplace=True)
df_batters
```

## Dropping some extreme outlier values

```python
def plot_bar(df, column):
    plt.figure(figsize=(12,8))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
initial_rows = df.shape[0]
before_rows = df.shape[0]

df = df[df['release_speed'] >= 69]
after_rows = df.shape[0]
rows_dropped = before_rows - after_rows
percentage_dropped = (rows_dropped / initial_rows) * 100
print(f"Number of rows dropped: {rows_dropped}")
print(f"Percentage of rows dropped: {percentage_dropped:.2f}%")
print(f"remaining rows: {after_rows}")
print()
before_rows = df.shape[0]
df = df[df['release_speed'] <= 100]
after_rows = df.shape[0]
rows_dropped = before_rows - after_rows
percentage_dropped = (rows_dropped / initial_rows) * 100
print(f"Number of rows dropped: {rows_dropped}")
print(f"Percentage of rows dropped: {percentage_dropped:.2f}%")
print(f"remaining rows: {after_rows}")
```

```python
#df.to_csv('df_batters_iso.csv', index=False)
df.drop(['G_y', 'G_x'], axis=1, inplace=True)
def homerun(row):
    if row['events'] == 'home_run':
        return 1
    else:
        return 0

df['home_run_hit'] = df.apply(homerun, axis=1)

df['home_run_in_game'] = df.groupby(['game_date', 'batter'])['home_run_hit'].transform('max')
df.to_csv('df_batters_iso.csv', index=False)
```
