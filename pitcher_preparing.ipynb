```python
def plot_bar(df, column):
    plt.figure(figsize=(12,8))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
master = People[['playerID', 'birthYear', 'nameFirst', 'nameLast', 'weight', 'height', 'debut']].copy()
# drop months and days from 'debut'
master['debut'] = master['debut'].str.slice(stop=4)
# configure names column as formatted in the primary data
master['name'] = master['nameLast'] + ', ' + master['nameFirst']
master = master.drop(['nameFirst', 'nameLast'], axis=1)
# drop the .0 ending from numerical values
master = master.dropna()
master['birthYear'] = master['birthYear'].astype(int)
master['weight'] = master['weight'].astype(int)
master['height'] = master['height'].astype(int)
master
```

```python
pitcher_stats = pd.merge(master, Pitching, on='playerID')
columns_to_drop = ['stint', 'lgID', 'W', 'L', 'GS', 'CG', 'SHO', 'SV', 'IPouts', 'ER', 'ERA', 'GF', 'SH', 'SF', 'GIDP']
pitcher_stats.drop(columns_to_drop, axis=1, inplace=True)
pitcher_stats = pd.merge(pitcher_stats, Teams, on=['yearID', 'teamID'], how='inner', suffixes=('_ps', '_teams'))
columns_to_drop = ['lgID', 'franchID', 'divID', 'Rank', 'Ghome', 'W', 'L', 'DivWin', 'WCWin', 'LgWin', 'WSWin', 'AB', 'H_teams', '2B', '3B', 'HR_teams', 'BB_teams', 'SO_teams', 'SB', 'CS', 'HBP_teams', 'SF', 'ER', 'ERA', 'CG', 'SHO', 'SV', 'IPouts', 'HA', 'HRA', 'BBA', 'SOA', 'park', 'attendance', 'BPF', 'PPF', 'teamIDBR', 'teamIDlahman45', 'teamIDretro']
pitcher_stats = pitcher_stats.drop(columns=columns_to_drop)
pitcher_stats = pitcher_stats.rename(columns={'BB_ps': 'Walks', 'IBB': 'Intentional_Walks', 'WP': 'Wild_Pitches', 'HBP_ps': 'Batters_Hit', 'BK': 'Balks', 'BFP': 'Batters_Faced'})
pitcher_stats
```

```python
df_pitchers = pd.merge(df, pitcher_stats, left_on=['player_name', 'game_year'], right_on=['name_ps', 'yearID'], how='inner')
df_pitchers
```

```python
null_columns = df_pitchers.columns[df_pitchers.isnull().any()]
null_counts = df_pitchers[null_columns].isnull().sum()
null_data = pd.DataFrame({'Column': null_columns, 'Null Count': null_counts})
null_data
```

```python
df_pitchers = df_pitchers.dropna(subset=['BAOpp']) # dropping a few null values
# new feature for calculating pitcher age
df_pitchers['Age'] = df_pitchers['game_year'] - df_pitchers['birthYear']
df_pitchers['debut'] = df_pitchers['debut'].astype(int)
df_pitchers['Year_Vet'] = df_pitchers['game_year'] - df_pitchers['debut']
df_pitchers = df_pitchers.drop(['debut', 'name_ps', 'yearID', 'birthYear', 'Balks'], axis=1)
df_pitchers
```

```python
# calculating some averages allowed per game/season total
df_pitchers['Avg_Hits_PerGame'] = df_pitchers['H_ps'] / df_pitchers['G_ps']
df_pitchers['Avg_HR_PerGame'] = df_pitchers['HR_ps'] / df_pitchers['G_ps']
df_pitchers['Avg_Walks_PerGame'] = df_pitchers['Walks'] / df_pitchers['G_ps']
df_pitchers['Avg_Strikeouts_PerGame'] = df_pitchers['SO_ps'] / df_pitchers['G_ps']
df_pitchers['Avg_Intent_Walks_PerGame'] = df_pitchers['Intentional_Walks'] / df_pitchers['G_ps']
df_pitchers['Avg_Wild_Pitches_PerGame'] = df_pitchers['Wild_Pitches'] / df_pitchers['G_ps']
df_pitchers['Avg_Batters_Hit_PerGame'] = df_pitchers['Batters_Hit'] / df_pitchers['G_ps']
df_pitchers['Avg_Batters_Faced_PerSeason'] = df_pitchers['Batters_Faced'] / df_pitchers['G_ps']
df_pitchers['Avg_Runs_Allowed_PerGame'] = df_pitchers['R_ps'] / df_pitchers['G_ps']
df_pitchers.drop(columns=['playerID', 'BAOpp', 'G_ps', 'teamID', 'H_ps', 'HR_ps', 'Walks', 'SO_ps', 'Intentional_Walks', 'Wild_Pitches', 'Batters_Hit', 'Batters_Faced', 'R_ps', 'G_teams', 'R_teams', 'RA', 'E', 'DP', 'FP', 'name_teams'], inplace=True)
df_pitchers
```
