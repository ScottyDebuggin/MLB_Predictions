```python
cols_to_keep_from_pitchers_all_years = ['game_date', 'batter', 'pitcher', 'stand', 'sz_top', 'sz_bot', 'home_score', 'away_score', 'fielder_2', 
                                        'release_speed', 'release_pos_x', 'release_pos_z']

df_pitchers_vs_batters = df_pitchers_solo.merge(
    pitchers_all_years[cols_to_keep_from_pitchers_all_years],
    on=['game_date', 'release_speed', 'pitcher', 'release_pos_x', 'release_pos_z'], 
    how='left'
)
df_pitchers_vs_batters
```
