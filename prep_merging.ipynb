# Merging the datasets

```python
def load_and_merge_files(start_year, end_year):
    # Directories to look in
    dirs = ['batter', 'pitcher', 'catcher']

    # Short names for the directories
    names = {'batter': 'b', 'pitcher': 'p', 'catcher': 'c'}

    # Initialize a dictionary to store the merged DataFrames
    merged_dataframes = {}

    # Process each year
    for year in range(start_year, end_year + 1):
        for d in dirs:
            # Build the directory path
            dir_path = os.path.join('data', 'master', str(year), d)

            # Get the list of files in the directory
            files = os.listdir(dir_path)

            # Initialize a list to store the DataFrames for this category
            dfs = []

            # Process each file
            for i, file in enumerate(files, start=1):
                # Build the full file path
                file_path = os.path.join(dir_path, file)

                # Read the file into a DataFrame
                df = pd.read_csv(file_path)

                # Add the DataFrame to the list
                dfs.append(df)

            # Concatenate all the DataFrames into one
            merged_df = pd.concat(dfs, ignore_index=True)

            # Drop columns where all values are null
            merged_df = merged_df.dropna(axis=1, how='all')

            # Convert DataFrame to lowercase
            merged_df.columns = merged_df.columns.str.lower()
            merged_df = merged_df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

            # Create a key for the merged DataFrame
            key = f'{d}_{year}'

            # Store the merged DataFrame in the dictionary
            merged_dataframes[key] = merged_df

    # Return the dictionary of merged DataFrames
    return merged_dataframes

# Test the function
merged_dataframes = load_and_merge_files(2008, 2022)
```

```python
# Define a function to merge all the dataframes of the same category
def merge_all_years(merged_dataframes, category):
    # Get the keys for the DataFrames in the given category
    keys = [key for key in merged_dataframes.keys() if key.startswith(category)]

    # Get the DataFrames for the given category
    dfs = [merged_dataframes[key] for key in keys]

    # Concatenate all the DataFrames into one
    df = pd.concat(dfs, ignore_index=True)

    # Return the merged DataFrame
    return df

# Merge all years for each category
batters_all_years = merge_all_years(merged_dataframes, 'batter')
pitchers_all_years = merge_all_years(merged_dataframes, 'pitcher')
catchers_all_years = merge_all_years(merged_dataframes, 'catcher')
```
